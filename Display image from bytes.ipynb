```python
%matplotlib ipympl
import matplotlib.pyplot as plt
import numpy as np
from bitstring import BitArray, Bits

with open("display-code0000-315.data", "rb") as f:
    data = f.read()

assert len(data) == 2520

img = np.zeros((315*8,8))
bits = Bits(data)

for i, b in enumerate(bits):
    y = i // 8
    x = i % 8
    img[y,x] = 1 if b else 0

plt.imsave("test.png", img,cmap="gray")


```
